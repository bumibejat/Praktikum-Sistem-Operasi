```python
from os import getpid
from time import time, sleep
from multiprocessing import Pool, Process

tipe = "all"

def gan_gen(angka):
    if tipe == "sekuensial":
        print(angka, ",'-ID proses',", getpid())
    elif tipe == "multiprocess":
        print(angka,",\"'-ID proses',Rahasia\"")
    elif tipe == "pool":
        print(angka, ",'-ID proses',", getpid())
    else:
        print("parameter salah")
    sleep(1)
    
if __name__ == '__main__':
    print("Input proses awal : ")
    awal = int(input())
    
    print("Input proses akhir : ")
    akhir = int(input())
    
    # PEMROSESAN SEKUENSIAL
    print("\nPemrograman Sekuensial")
    tipe = "sekuensial"
    sekuensial_awal = time()
    
    for i in range(awal, akhir+1):
      gan_gen(i)
    
    sekuensial_akhir = time()
    
    # PEMROSESAN PARALEL DENGAN multiprocessing.process
    print("\nMultiprocessing.Process")
    tipe = "multiprocess"
    kumpulan_proses = []
    
    process_awal = time()
    
    for i in range(awal, akhir+1):
      p = Process(target = gan_gen, args = (i,))
      kumpulan_proses.append(p)
      p.start()
    
    for i in kumpulan_proses:
      p.join()
    
    process_akhir = time()
    
    # PEMROSESAN PARALEL DENGAN multiprocessing.pool
    tipe = "pool"
    print("\nMultiprocessing.Pool")
    pool_awal = time()
    
    pool = Pool()
    pool.map(gan_gen, range(awal , akhir+1))
    pool.close()
    
    pool_akhir = time()
    
print("\nSekuensial", sekuensial_akhir-sekuensial_awal, "detik")
print("Multiprocessing.Process", process_akhir-process_awal, "detik")
print("Multiprocessing.Pool", pool_akhir-pool_awal, "detik")
```
